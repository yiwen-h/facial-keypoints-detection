### Percentage of Missing Data in Column

```python
percent_nan_in_each_col = data.isna().sum()/(len(data))*100
print(percent_nan_in_each_col)
```

### Number of NaN Values in Each Row with reference to column

### Removing Rows Having Missing Data

```python
# remove all rows that have atleast one missing value
data.dropna(inplace=True)

# remove all rows that have all values missing
data.dropna(how='all',inplace=True)
```

```python
data.head
```

### Removing Columns Having Missing Data

```python
# remove all columns that have atleast one missing value
data.dropna(axis='columns',inplace=True)

# remove all columns that have all values missing
data.dropna(axis='columns',how='all',inplace=True)
```

```python
data.head
```

### Replace Missing Data with Mean Value

```python
# Replace all NaN value of Column 'left_eye_inner_corner_x' by Average of all points
mean_value =data['left_eye_inner_corner_x'].mean()
data['left_eye_inner_corner_x'].fillna(mean_value,inplace=True)
print(data)
```

```python
# Replace all NaN value of Column 'left_eye_inner_corner_y' by Average of all points
mean_value =data['left_eye_inner_corner_y'].mean()
data['left_eye_inner_corner_y'].fillna(mean_value,inplace=True)
print(data)
```
